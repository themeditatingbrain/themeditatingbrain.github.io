```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_context('poster')
```

```python
data1 = pd.read_csv('data/meditating brain 2019 - Andy.csv')
data2 = pd.read_csv('data/meditating brain 2019 - Steven.csv')
data = pd.concat([data1, data2], axis = 0).dropna()
data = data[~((data['Time']=='After') & (data['Meditation']=='Retreat'))]
data = data[data['Meditation']!='Memory']
data.head()
```

```python
def plotQuestion(data, question, ylim):
    fig = plt.figure(figsize = (10, 8))
    ax = fig.add_subplot(1, 1, 1)
    sns.barplot(data = data[data["Question"] == question], x = 'Meditation', y = 'Rating', hue = 'Time', ci = 68, ax = ax)
    plt.title(question)
    plt.ylim(ylim[0], ylim[1])
    plt.savefig('img/' + question.lower().replace('/', '') + '-2019.png')
    plt.show()
```

```python
plotQuestion(data, 'Calm/Relaxed', (5, 10))
```

```python
plotQuestion(data, 'Stressed/Anxious', (0, 5))
```

```python
plotQuestion(data, 'Focused/Attentive', (5, 8))
```

```python
plotQuestion(data, 'Open/Aware', (5, 10))
```

```python
plotQuestion(data, 'Pleasant Emotional State', (5, 10))
```
